```python
#creating months and each month max days , which will be used to map each city record on particular day 
months_list=['Jan','Feb','March','April','May','June','July','August','Sept','Nov','Oct','Dec']
days_in_each_month=[31,28,31,30,31,30,31,31,30,31,30,31]
```

```python
#making matrix for all months for each cities
#for each category as well
cities_matrix_confirmed=np.zeros((len(city_names),12,31))
cities_matrix_deaths=np.zeros((len(city_names),12,31))
cities_matrix_recovered=np.zeros((len(city_names),12,31))
```

```python
#now searching in the whole exel sheets , and updating the array as with it continuously
my_sum=0
for single_row in arrray:
   
    
    time_stamp=single_row[0]
    time_stamp=pd.to_datetime(time_stamp, format='%m/%d/%Y')
    print(time_stamp)
    entry_month=time_stamp.month
    entry_day=time_stamp.day
    if type(entry_day)!=type(1) and type(entry_month)!=type(1):
        continue
    
    #then it needs to update the mark the entry in respective date counter
    
    #also now find the particular district and add counter in there as well
        
    extracted_district=single_row[6]
    print(extracted_district)
    extracted_district=extracted_district.strip()
    
    if extracted_district!='NSB':
    
        extracted_district=extracted_district.title()
    
#         print(extracted_district)
    index_of_district=city_names.index(extracted_district)
    cities_matrix_confirmed[index_of_district,entry_month-1,entry_day-1]=single_row[1]+cities_matrix_confirmed[index_of_district,entry_month-1,entry_day-1]
    my_sum=my_sum+single_row[1]
    cities_matrix_deaths[index_of_district,entry_month-1,entry_day-1]=single_row[2]+cities_matrix_deaths[index_of_district,entry_month-1,entry_day-1]
    
    cities_matrix_recovered[index_of_district,entry_month-1,entry_day-1]=single_row[3]+cities_matrix_recovered[index_of_district,entry_month-1,entry_day-1]
        
```

```python
#the three variables cities_matrix_confirmed,
#cities_matrix_deaths and cities_matrix_recovered contains all the cases citywise wrt each day and each month
#their size is (cities x 12 x31 ) , where 12 is months dimension,
#31 days dimension and cities are no of cities whose records are available
```

```python
#now plotting the results 
def my_plot(month,till_date):
    #given month number , and max data length ,
    #it will plot the total cases , total recoveries and total deaths in pakistan so far
    squeezed_confrimed=(np.sum(cities_matrix_confirmed,axis=0))
    squeezed_recovered=(np.sum(cities_matrix_recovered,axis=0))
    squeezed_death=(np.sum(cities_matrix_deaths,axis=0))
    plt.plot(range(1,till_date+1),np.cumsum(squeezed_confrimed[month-1,0:till_date]))
    plt.xlabel('Days of Month :' +str(months_list[month-1]))
    plt.ylabel('No of confirmed Cases')
    plt.title('POSITIVE CASES PER DAY')
    plt.show()
    plt.plot(range(1,till_date+1),np.cumsum(squeezed_recovered[month-1,0:till_date]))
    plt.xlabel('Days of Month :' +str(months_list[month-1]))
    plt.ylabel('No of RECOVERIES')
    plt.title('RECOVERY PER DAY ')
    plt.show()
    plt.plot(range(1,till_date+1),np.cumsum(squeezed_death[month-1,0:till_date]))
    plt.xlabel('Days of Month :' +str(months_list[month-1]))
    plt.ylabel('No of Deaths')
    plt.title('DEATHS PER DAY')
    plt.show()
#calling the function
my_plot(4,30)
```

```python
#NOW CITIWISE DATA VISUALIZATION 
#another function which will plot the records citywise , given the cases type , max date,
#top_N cities wrt to maximum cases and month number 
def plot_district_graphs(type_of_case,till_date,top_numbers,month_no):
    if type_of_case=='positive':
            summed_arrays_districts=np.sum(cities_matrix_confirmed[:,month_no-1,:],axis=1)
            choosen_type=cities_matrix_confirmed
            top_N_district_indices=sorted(range(len(summed_arrays_districts)), key=lambda i: summed_arrays_districts[i], reverse=True)[:top_numbers]
    elif type_of_case=='recovery':
            summed_arrays_districts=np.sum(cities_matrix_recovered[:,month_no-1,:],axis=1)
            choosen_type=cities_matrix_recovered
            top_N_district_indices=sorted(range(len(summed_arrays_districts)), key=lambda i: summed_arrays_districts[i], reverse=True)[:top_numbers]
    else:
            summed_arrays_districts=np.sum(cities_matrix_deaths[:,month_no-1,:],axis=1)
            choosen_type=cities_matrix_deaths
            top_N_district_indices=sorted(range(len(summed_arrays_districts)), key=lambda i: summed_arrays_districts[i], reverse=True)[:top_numbers]
    
    
    print(top_N_district_indices)
    for i in top_N_district_indices:
        x=range(1,(till_date)+1)
        
        y1=choosen_type[i,month_no-1,0:till_date]   
        y=np.cumsum(y1)
        print(np.sum(y1))
        plt.plot(x,y)
        plt.xlabel('MONTH='+str(months_list[month_no-1])+' (DAYS_AXIS)')
        plt.ylabel(type_of_case+'  CASES')
        plt.title('City:'+ city_names[i]+'--- total cases = '+ str(np.sum(y1)))
        plt.show()
```

```python
#this line will plot positive cases of top 5 cities (wrt maximum cases) 
#of month april (4) with a max date of 30 days
plot_district_graphs('positive',30,15,4)
```

```python
#making matrix for single month for each district (FOR TABLEU DISTRICTS)
#for each category as well 
districts_matrix_confirmed_tableau=np.zeros((len(my_list),12,31))
districts_matrix_deaths_tableau=np.zeros((len(my_list),12,31))
districts_matrix_recovered_tableau=np.zeros((len(my_list),12,31))
```

```python
#Now extracting the data to form district wise 3 matrix variables defined above
#again the size of those matrix are (districts x months x days) --> so we will have district wise analysis

counter=0
i=0
my_sum=0
for index in range(x.shape[1]):
    #first for confirmed cases
    temp_array_positive=np.zeros((1,12,31))
    temp_array_recovered=np.zeros((1,12,31))
    temp_array_deaths=np.zeros((1,12,31))
    for columns in range(1,(x.shape[0])):
        
        if (pd.isnull(x[columns,index])==True or x[columns,index]=='not_found'):
            break
        else:
#             counter=counter+1
#             print(counter)
            #now add and store in above saved districts matrix
            find_index=city_names.index(x[columns,index])
            print(x[columns,index])
            i=i+1
#             print(find_index)
            my_sum=my_sum+np.sum(districts_matrix_confirmed[find_index,:,:])
            temp_array_positive=temp_array_positive+cities_matrix_confirmed[find_index,:,:]
            temp_array_recovered=temp_array_recovered+cities_matrix_recovered[find_index,:,:]
            temp_array_deaths=temp_array_deaths+cities_matrix_deaths[find_index,:,:]
            
#     print(temp_array.shape)
    districts_matrix_confirmed_tableau[index,:,:]=temp_array_positive
    districts_matrix_recovered_tableau[index,:,:]=temp_array_recovered
    districts_matrix_deaths_tableau[index,:,:]=temp_array_deaths
    
print(i)
print(my_sum)
```
