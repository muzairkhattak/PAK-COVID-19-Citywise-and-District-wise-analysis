```python
cases_confirmed=np.zeros((12,31))
cases_suspected=np.zeros((12,31))
cases_probable=np.zeros((12,31))
```

```python
months_list=['Jan','Feb','March','April','May','June','July','August','Sept','Nov','Oct','Dec']
```

```python
#converting counted_data into dictionary 
all_cases_dict={'confirmed' : cases_confirmed,
                'suspected' : cases_suspected,
                'probable'  : cases_probable
               }
```

```python
district_read_array=np.array(district_read)
```

```python
district_names=district_read_array.reshape((district_read_array.shape[0])*(district_read_array.shape[1]))
```

```python
district_names=list(district_names)
del(district_names[-1])
```

```python
#making matrix for single month for each district
#for each category as well
districts_matrix_suspected=np.zeros((len(district_names),31))
districts_matrix_confirmed=np.zeros((len(district_names),31))
districts_matrix_probable=np.zeros((len(district_names),31))
```

```python
#now searching in the whole exel sheets , and updating the array as with it continuously
for single_row in arrray:
    time_stamp=single_row[3]
    time_stamp.to_pydatetime()
    entry_month=time_stamp.month
    entry_day=time_stamp.day
    if type(entry_day)!=type(1) and type(entry_month)!=type(1):
        continue
    #now time to see if case is confirm or not
    case_type=single_row[21]
    if case_type=='confirmed':
        #then it needs to update the mark the entry in respective date counter
        cases_confirmed[entry_month-1,entry_day-1]=cases_confirmed[entry_month-1,entry_day-1]+1
        #also now find the particular district and add counter in there as well
        
        extracted_district=single_row[58]
#         print(extracted_district)
        index_of_district=district_names.index(extracted_district)
        districts_matrix_confirmed[index_of_district,entry_day-1]=districts_matrix_confirmed[index_of_district,entry_day-1]+1
        
    elif case_type=="suspected":
       
        #then it needs to update the mark the entry in respective date counter
        cases_suspected[entry_month-1,entry_day-1]=cases_suspected[entry_month-1,entry_day-1]+1
        #also now find the particular district and add counter in there as well
        extracted_district=single_row[58]
#         print(extracted_district)
        index_of_district=district_names.index(extracted_district)
        districts_matrix_suspected[index_of_district,entry_day-1]=districts_matrix_suspected[index_of_district,entry_day-1]+1
        
    else:
        cases_probable[entry_month-1,entry_day-1]=cases_probable[entry_month-1,entry_day-1]+1
        
        #also now find the particular district and add counter in there as well
        extracted_district=single_row[58]
#         print(extracted_district)
        index_of_district=district_names.index(extracted_district)
        districts_matrix_probable[index_of_district,entry_day-1]=districts_matrix_probable[index_of_district,entry_day-1]+1
```

```python
#Now fetching data from filled_matrices
my_bin=[]

my_frequencies=None
def make_hist_case_type(month_name,case_type):
    my_bin=[]
    for i in range(31):
        my_bin.append(month_name+ str(i+1))
    my_bin=np.array(my_bin)
#     my_bin=my_bin+1
    cases_matrix=all_cases_dict[case_type]
    matrix_index=months_list.index(month_name)
    my_frequencies=cases_matrix[matrix_index,:]
    plt.figure(figsize=(20,10))
    index = np.arange(len(my_bin))
#     plt.figure(figsize=(20, 3))  # width:20, height:3
    plt.bar(index, my_frequencies, align='edge', width=0.9)
    plt.xticks(index, my_bin, fontsize=15, rotation=30)
    plt.title((case_type+' Cases'),fontsize=20)
    plt.ylabel('Frequency',fontsize=20)
    plt.xlabel('Days',fontsize=20)
    
#     print(my_frequencies.shape)
#     print(my_bin.shape)

```

```python
make_hist_case_type('March','confirmed')
plt.show()
make_hist_case_type('March','suspected')
plt.show()
make_hist_case_type('March','probable')
```

```python
#now time to make the graphs
#single month , and each district


def plot_district_graphs(type_of_case,till_date,top_numbers):
    if type_of_case=='confirmed':
            summed_arrays_districts=np.sum(districts_matrix_confirmed,axis=1)
            choosen_type=districts_matrix_confirmed
            top_N_district_indices=sorted(range(len(summed_arrays_districts)), key=lambda i: summed_arrays_districts[i], reverse=True)[:top_numbers]
    elif type_of_case=='suspected':
            summed_arrays_districts=np.sum(districts_matrix_suspected,axis=1)
            choosen_type=districts_matrix_suspected
            top_N_district_indices=sorted(range(len(summed_arrays_districts)), key=lambda i: summed_arrays_districts[i], reverse=True)[:top_numbers]
    else:
            summed_arrays_districts=np.sum(districts_matrix_probable,axis=1)
            choosen_type=districts_matrix_probable
            top_N_district_indices=sorted(range(len(summed_arrays_districts)), key=lambda i: summed_arrays_districts[i], reverse=True)[:top_numbers]
    
    
    
    for i in top_N_district_indices:
        x=range(1,(till_date)+1)
        
        y1=choosen_type[i,0:till_date]   
        y=np.cumsum(y1)
        plt.plot(x,y)
        plt.xlabel('MONTH= MARCH (DAYS_AXIS)')
        plt.ylabel(type_of_case+'  CASES')
        plt.title('District:'+ district_names[i]+'--- total cases = '+ str(np.sum(y1)))
        plt.show()
        

```
