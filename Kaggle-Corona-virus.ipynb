```python
#creating months and each month max days , which will be used to map each city record on particular day 
months_list=['Jan','Feb','March','April','May','June','July','August','Sept','Nov','Oct','Dec']
days_in_each_month=[31,28,31,30,31,30,31,31,30,31,30,31]
```

```python
#making matrix for all months for each cities
#for each category as well
districts_matrix_confirmed=np.zeros((len(district_names),12,31))
districts_matrix_deaths=np.zeros((len(district_names),12,31))
districts_matrix_recovered=np.zeros((len(district_names),12,31))
```

```python
#now searching in the whole exel sheets , and updating the array as with it continuously
my_sum=0
for single_row in arrray:
   
    
    time_stamp=single_row[0]
    time_stamp=pd.to_datetime(time_stamp, format='%m/%d/%Y')
    print(time_stamp)
    entry_month=time_stamp.month
    entry_day=time_stamp.day
    if type(entry_day)!=type(1) and type(entry_month)!=type(1):
        continue
    
    #then it needs to update the mark the entry in respective date counter
    
    #also now find the particular district and add counter in there as well
        
    extracted_district=single_row[6]
    print(extracted_district)
    extracted_district=extracted_district.strip()
    
    if extracted_district!='NSB':
    
        extracted_district=extracted_district.title()
    
#         print(extracted_district)
    index_of_district=district_names.index(extracted_district)
    districts_matrix_confirmed[index_of_district,entry_month-1,entry_day-1]=single_row[1]+districts_matrix_confirmed[index_of_district,entry_month-1,entry_day-1]
    my_sum=my_sum+single_row[1]
    districts_matrix_deaths[index_of_district,entry_month-1,entry_day-1]=single_row[2]+districts_matrix_deaths[index_of_district,entry_month-1,entry_day-1]
    
    districts_matrix_recovered[index_of_district,entry_month-1,entry_day-1]=single_row[3]+districts_matrix_recovered[index_of_district,entry_month-1,entry_day-1]
        
```

```python
#now plotting the results
def my_plot(month,till_date):
    
    squeezed_confrimed=(np.sum(districts_matrix_confirmed,axis=0))
    squeezed_recovered=(np.sum(districts_matrix_recovered,axis=0))
    squeezed_death=(np.sum(districts_matrix_deaths,axis=0))
    plt.plot(range(1,till_date+1),np.cumsum(squeezed_confrimed[month-1,0:till_date]))
    plt.xlabel('Days of Month :' +str(months_list[month-1]))
    plt.ylabel('No of confirmed Cases')
    plt.title('POSITIVE CASES PER DAY')
    plt.show()
    plt.plot(range(1,till_date+1),np.cumsum(squeezed_recovered[month-1,0:till_date]))
    plt.xlabel('Days of Month :' +str(months_list[month-1]))
    plt.ylabel('No of RECOVERIES')
    plt.title('RECOVERY PER DAY ')
    plt.show()
    plt.plot(range(1,till_date+1),np.cumsum(squeezed_death[month-1,0:till_date]))
    plt.xlabel('Days of Month :' +str(months_list[month-1]))
    plt.ylabel('No of Deaths')
    plt.title('DEATHS PER DAY')
    plt.show()
my_plot(5,10)
```

```python
def plot_district_graphs(type_of_case,till_date,top_numbers,month_no):
    if type_of_case=='positive':
            summed_arrays_districts=np.sum(districts_matrix_confirmed[:,month_no-1,:],axis=1)
            choosen_type=districts_matrix_confirmed
            top_N_district_indices=sorted(range(len(summed_arrays_districts)), key=lambda i: summed_arrays_districts[i], reverse=True)[:top_numbers]
    elif type_of_case=='recovery':
            summed_arrays_districts=np.sum(districts_matrix_recovered[:,month_no-1,:],axis=1)
            choosen_type=districts_matrix_recovered
            top_N_district_indices=sorted(range(len(summed_arrays_districts)), key=lambda i: summed_arrays_districts[i], reverse=True)[:top_numbers]
    else:
            summed_arrays_districts=np.sum(districts_matrix_deaths[:,month_no-1,:],axis=1)
            choosen_type=districts_matrix_deaths
            top_N_district_indices=sorted(range(len(summed_arrays_districts)), key=lambda i: summed_arrays_districts[i], reverse=True)[:top_numbers]
    
    
    print(top_N_district_indices)
    for i in top_N_district_indices:
        x=range(1,(till_date)+1)
        
        y1=choosen_type[i,month_no-1,0:till_date]   
        y=np.cumsum(y1)
        print(np.sum(y1))
        plt.plot(x,y)
        plt.xlabel('MONTH='+str(months_list[month_no-1])+' (DAYS_AXIS)')
        plt.ylabel(type_of_case+'  CASES')
        plt.title('City:'+ district_names[i]+'--- total cases = '+ str(np.sum(y1)))
        plt.show()
```

```python
#NOW 16/04/2020
#FOR TIME SERIES-DATA-COMMULATIVE
#1013 reading because we were failed to map 4 disgrict readings from code for pakistan into the tablue districts list
#finally _exporting the data to Ali bhai for mapping in the map of actual counts of covid cases
#Now exporting the ordered data (time-series) into excel file for analysis

months=[2,3,4,5,6]


cummulative_death_perday_cases=[]
cummulative_recovery_perday_cases=[]
cummulative_perday_cases=[]
# state=list(np.array(tablu_districts['province_territory']))
# print(state[50])
Day_number=[]
my_date_list=[]
#exporting results for confirmed cases first
my_list_for_timeseries=[]
day=1
counter=1
till_date_xoxo=3
for i,district in enumerate(my_list):
    day=1
    recovery_last_month_sum=0
    deaths_last_month_sum=0
    confirmed_last_month_sum=0
    #now district is fixed , iterate same district for all given months
    for  month in months:
        current_district_cummulative_recovery=np.cumsum(districts_matrix_recovered_tableau[i,month-1,:])+recovery_last_month_sum
        current_district_cummulative_deaths=np.cumsum(districts_matrix_deaths_tableau[i,month-1,:])+deaths_last_month_sum
        current_district_cummulative=np.cumsum(districts_matrix_confirmed_tableau[i,month-1,:])+confirmed_last_month_sum
        recovery_last_month_sum=0
        deaths_last_month_sum=0
        confirmed_last_month_sum=0
        recovery_last_month_sum=current_district_cummulative_recovery[-1]+recovery_last_month_sum
        deaths_last_month_sum=current_district_cummulative_deaths[-1]+deaths_last_month_sum
        confirmed_last_month_sum=current_district_cummulative[-1]+confirmed_last_month_sum
        
        for each_day in range(len(districts_matrix_confirmed_tableau[i,month-1,0:days_in_each_month[month-1]])):
            Day_number.append(day)
            day=day+1
            my_date_list.append(str(each_day+1) +'/' +'0'+ str(month)+'/2020')
            my_list_for_timeseries.append(district)
            cummulative_recovery_perday_cases.append(current_district_cummulative_recovery[each_day])
            cummulative_death_perday_cases.append(current_district_cummulative_deaths[each_day])
            cummulative_perday_cases.append(current_district_cummulative[each_day])
            if (each_day==till_date_xoxo-1 and month==months[-1]):
                print(counter)
                counter=counter+1
                break
df={'Day No': Day_number,'Date':my_date_list,'District':my_list_for_timeseries,'positive_cases':cummulative_perday_cases,'recovered_cases':cummulative_recovery_perday_cases,'deaths_cases':cummulative_death_perday_cases}
my_csv=pd.DataFrame(df)
my_csv.to_csv(r'C:\Users\mkhat\Desktop\from_kaggle_covid_data\time_series_3rd_june.csv')
```
